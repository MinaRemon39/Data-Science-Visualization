```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## **Trends** 

**1. lineplot(data):** are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group. x= index y= columns values

```python
fifa=pd.read_csv("Data/fifa.csv",index_col="Date",parse_dates=True) # Read Data

plt.subplots(figsize=(16,5)) # Size of plot (width,height)

sns.lineplot(data=fifa)

plt.show()
```

## **Relationship** 

**1. barplot(data,x,y,hue):** useful for comparing quantities corresponding to different groups

```python
ign=pd.read_csv("Data/ign_scores.csv") 

plt.subplots(figsize=(30,8))

sns.barplot(data=ign,x="Platform",y="Racing")

plt.show()
```

**2. heatmap(data,annot):** can be used to find color-coded patterns in tables of numbers.

```python
ign.set_index("Platform",inplace=True)

plt.subplots(figsize=(12,8))

sns.heatmap(data=ign,annot=True) # annot = true to show the values

plt.show()
```

**3. scatterplot(data,x,y,hue):** show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third categorical variable.

```python
insurance = pd.read_csv("Data/insurance.csv")

sns.scatterplot(data=insurance,x="bmi",y="charges",hue="smoker")

plt.show()
```

**4. regplot(data,x,y):** including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.

```python
sns.regplot(data=insurance,x="bmi",y="charges")

plt.show()
```

**5. swarmplot(data,x,y):** Categorical scatter plots show the relationship between a continuous variable and a categorical variable.

```python
sns.swarmplot(data=insurance,x="smoker",y="charges")

plt.show()
```

**6. lmplot(data,x,y,hue):** This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.

```python
sns.lmplot(data=insurance,x="bmi",y='charges',hue="smoker")

plt.show()
```

## **Distribution**

**1. histplot(data,x,hue,kde):** show distripution for 1 cloumn , show realtion between 1 cloumn and its count

```python
cancer=pd.read_csv("Data/cancer.csv")

sns.histplot(data=cancer,x="Area (mean)",hue="Diagnosis")

plt.show()
```

**2. kdeplot(data,x,hue,shade):** show smoth distripution for 1 column , show relation between 1 column and its density ( smoth histogram ) 

```python
sns.kdeplot(data=cancer,x="Area (mean)",hue="Diagnosis",shade=True)
```

**3. jointplot(data,x,y,kind="kde"):** show 2d kde relationship between 2 columns 

```python
sns.jointplot(data=cancer,x="Area (mean)",y="Perimeter (mean)",kind="kde")
```
